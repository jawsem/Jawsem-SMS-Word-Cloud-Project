## Creating a Text Message Word Cloud

1. Create a sqllite database to store all my text data.

## 3. Extract information out of the data base necessary to create a word cloud.

**I pretty much already have thid done since all the data is stored in my dataframe**

```python
df[df['contact_name'] == 'Kaiti Pie']['type'].value_counts()
df['contact_name'].unique()
```

```python

df['received'] = df['type'].apply(lambda x: 1 if x == '1' else 0)
df['sent'] = df['type'].apply(lambda x: 1 if x == '2' else 0)

        

```

```python
plt.style.use('ggplot')
df['readable_date'] = pd.to_datetime(df['readable_date'])

def received_vs_sent(contact):
    fig1, ax1 = plt.subplots(figsize=(10,10))
    x1 = df[(df['contact_name'] == contact) & (df['type'] == '1')]['readable_date']
    x2 = df[(df['contact_name'] == contact) & (df['type'] == '2')]['readable_date']
    y1 = df[(df['contact_name'] == contact) & (df['type'] == '1')]['received'].cumsum()
    y2 = df[(df['contact_name'] == contact) & (df['type'] == '2')]['sent'].cumsum()

    ax1.plot(x1,y1)
    ax1.plot(x2,y2)
    handles, labels = ax1.get_legend_handles_labels()
    ax1.legend(handles, labels)
    ax1.set_title('Texts with {}'.format(contact))
    plt.show()
    return fig1, ax1


received_vs_sent('Kaiti Pie');
```

```python
word_cloud_contacts = []
for contact in df['contact_name'].unique():
    if df[df['contact_name'] ==contact].shape[0] > 100:
        word_cloud_contacts.append(contact)
        received_vs_sent(contact);
```

**Intesting metrics for the my most texted individuals.  I can see from the graphs above how many texts we have accumulated together as well as a comparison between the received and sent texts**

## 4. Create the word cloud of all words in the text messages between my wife and I.

**Now to create the word clouds.  I am starting with this [link](https://www.geeksforgeeks.org/generating-word-cloud-python/) as the template to start creating them and I am going to edit that as I go along.**

```python
##import the word cloud module
from wordcloud import WordCloud, STOPWORDS

##The goal here is to graph a word cloud for each of the most texted individuals
stopwords = set(STOPWORDS)

for contact in word_cloud_contacts:
    comments_words = ' '
    for body in df[df['contact_name'] == contact]['body']:
        body = str(body)
        tokens = body.split()
        for i in range(len(tokens)):
            tokens[i] = tokens[i].lower()
        for words in tokens:
            comments_words = comments_words + words + ' '

    wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comments_words)
 
    # plot the WordCloud image                       
    plt.figure(figsize = (8, 8), facecolor = None)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.tight_layout(pad = 0)
    plt.title("{} Word Cloud".format(contact))
    plt.show()
```

```python
##Creating a function to generate the word clouds

def generate_word_cloud(contact, stopwords, imp_words = [], imp_factor = 2, max_words = 200):
#     Param: contact(str) = contact you want to create a wordcloud for
#             stopwords(set) = set stop words that you dont want to count
#             imp_words(list) = list of words that you want to count more, or dict if dict is used we ignore imp_factor
#             imp_factor(number) = number of times you want to multiply those words
            
#     Returns:
#             wordcloud = wordcloud object 
#             ax = ax that your plot is on
#             fig = fig that your plot is on

    comments_words = ' '
    for body in df[df['contact_name'] == contact]['body']:
        body = str(body)
        tokens = body.split()
        for i in range(len(tokens)):
            tokens[i] = tokens[i].lower()
        for words in tokens:
            if type(imp_words) == dict:
                if words in imp_words.keys():
                    words = (words+' ')*imp_words[words]
                else:
                    words = words+' '
            elif type(imp_words) == list:
                if words in imp_words:
                    words = (words+' ')*imp_factor
                else:
                    words = words + ' '
            comments_words = comments_words + words

    wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                collocations = False,
                max_words = max_words,
                min_font_size = 10).generate(comments_words)
    fig, ax = plt.subplots(figsize = (8,8), facecolor = None)
    ax.imshow(wordcloud)
    ax.axis("off")
    plt.tight_layout(pad = 0)
    plt.show()
    return wordcloud, ax, fig, comments_words


```

```python
imp_words = ['oliver', 'ryder', 'dog','cutie']
imp_words_dict = {'oliver':4,'ryder':4, 'dog':3, 'cutie':3}
stopwords = set(STOPWORDS)
word_list = ['ok','https imgur','now','going','got','go','gifv','im','oh','lot','yet','bad','ill','gifv','https','https instagram','imgur','haha','hahaha','today','think','day',
            'want','good','well','work','still','one','done','will','time','back','right','see','getting','really',
            'facebook', 'netflix', 'wiki']

stopwords.update(word_list)
wordcloud, ax, fig, comments_words = generate_word_cloud('Kaiti Pie', stopwords, imp_words_dict, max_words = 100)

```

**We have been messing around with displaying the image a lot in the past few cells.  In order to actually do it correctly, I am going to define what we need to do**

1. We need to take an existing image and put it over a white background that is slightly larger.
2. We need to convert the image to be black so we can use it as a mask for our word cloud.
3. We need to create the word cloud on the border.
4. We need to save that word cloud as an image object.
5. We need to paste the actual image over the word cloud so it is centered within the word cloud.

```python

##Create initial Image array
image_array = np.array(Image.open('new_kaiti_dogs.jpg')).transpose((1,0,2))
width = int(image_array.shape[0]*1.2)
height = int(image_array.shape[1]*1.2)
##create the white background
background_image = Image.fromarray((np.ones((width,height,image_array.shape[2]),np.uint8))*255,'RGB')
##get image object
image = Image.fromarray(image_array,'RGB')

##Set position to paste
position = ((int(round(background_image.width - image.width)/2)),int(round(background_image.height-image.height)/2))

##Paste it, save and display the images
background_image.paste(image, position)
background_image.save('test/a_third_test.png')
display(Image.open('test/a_third_test.png'))

##Create the mask
mask = np.array(background_image)

##Set up the transformed mask
transformed_mask = np.ndarray((mask.shape[0],mask.shape[1], mask.shape[2]), np.int32)

##Create the white array
white_array = np.array([255,255,255])

##Alter the mask so all the white is black and all the images are white
for i in range(len(mask)):
    transformed_mask[i] = list(map(lambda x: [0,0,0] if ((x == white_array).sum() ==3) else white_array, mask[i]))

print(transformed_mask[0][2], transformed_mask[0][320])
##Save the mask as an image
img = Image.fromarray(transformed_mask, 'RGB')
img.save('test/a_fourth_test.png')
img.show()

##Create the word cloud with said mask
kaiti_cloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                collocations = False,
                max_words = 100,
                mask = transformed_mask,
                min_font_size = 10,
                contour_width = 3,
                 contour_color = 'green').generate(comments_words)

##Save the wordcloud to a file
kaiti_cloud.to_file('test/a_fifth_test.png')
cloud_image = Image.open('test/a_fifth_test.png')
display(cloud_image)
new_position = ((int(round(cloud_image.width - image.width)/2)),int(round(cloud_image.height-image.height)/2))
cloud_image.paste(image,new_position)

cloud_image.save('test/final_product.png')
display(cloud_image)
```

```python
##Lets create a function to do this.

def border_wordcloud(words, img, background_color = 'white', border_size = 1.2,stopwords = set(STOPWORDS), can_display = False):
#     Takes in words and an image and displays the word cloud in a border around the image
#     Params
#         words: list of all the words you supplying to the word cloud
#         image: the image you wish to display your word cloud around
    
#     Returns
#         cloud_image: the image with the word cloud around it

#     Create initial Image array
    image_array = np.array(img)
    
    width = int(image_array.shape[0]*border_size)
    height = int(image_array.shape[1]*border_size)
    
    ##create the white background
    background_image = Image.fromarray((np.ones((width,height,image_array.shape[2]),np.uint8))*255,'RGB')
    
    ##Set position to paste
    position = ((int(round(background_image.width - img.width)/2)),int(round(background_image.height-img.height)/2))

    ##Paste it, save and display the images
    background_image.paste(img, position)
    
    ##Create the mask
    mask = np.array(background_image)

    ##Set up the transformed mask
    transformed_mask = np.ndarray((mask.shape[0],mask.shape[1], mask.shape[2]), np.int32)

    ##Create the white array
    white_array = np.array([255,255,255])

    ##Alter the mask so all the white is black and all the images are white
    for i in range(len(mask)):
        transformed_mask[i] = list(map(lambda x: [0,0,0] if ((x == white_array).sum() ==3) else white_array, mask[i]))
    
    ##Create the word cloud with said mask
    word_cloud = WordCloud(width = 800, height = 800,
                    background_color = background_color,
                    stopwords = stopwords,
                    collocations = False,
                    max_words = 100,
                    mask = transformed_mask,
                    min_font_size = 10,
                    contour_width = 3,
                     contour_color = 'green').generate(words)

    ##Save the wordcloud to a file
    cloud_image = word_cloud.to_image()
    new_position = ((int(round(cloud_image.width - img.width)/2)),int(round(cloud_image.height-img.height)/2))
    cloud_image.paste(img,new_position)
    if can_display:
        display(cloud_image)
    return cloud_image
```
